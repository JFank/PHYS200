```python
import matplotlib.pyplot as pltimport numpy as np
```

```python
x = np.linspace(0,2*np.pi)y = np.sin(x)plt.plot(x,y,'r--',label = 'sin(x)')plt.legend()plt.grid()plt.title('Harmonic')plt.xlabel(r'$\pi$')plt.ylabel('y')
```

```python
x = np.arange(.1,4,.05)y = np.exp(-x)yerr = .1 + .2*np.sqrt(x)xerr = .1 + yerrplt.figure()plt.errorbar(x,y,xerr=xerr,yerr=yerr)plt.title('errorbars')
```
